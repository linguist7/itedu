### **| 변수 요약**

* 속성을 그리기전 bool 값을 가지는 속성 `adult_mal`과 `alone`을 int 값으로 변환해야 합니다.

## 데이터 분포
---

### **| 데이터 분포 시각화**

* 연속 자료형 분포 그리기

```python
fig, axes = plt.subplots(1, len(nums), figsize=(10,4))
axes = axes.flatten()
for col, ax in zip(nums, axes):
  sns.histplot(data=df, x=col, ax=ax)
plt.show()
```

* 이산 자료형 분포 그리기

```python
fig, axes = plt.subplots(2, 6, figsize=(24,10))
axes = axes.flatten()
for col, ax in zip(noms, axes):
  sns.countplot(data=df, x=col, ax=ax)
plt.show()
```

### | **기술 통계량 시각화**

* 기술통계량 확인

```python
df.describe()
```

* boxplot 으로 연속 자료형 기술 통계 시각화하기

```python
fig, axes = plt.subplots(1, len(nums), figsize=(10,4))
axes = axes.flatten()
for col, ax in zip(nums, axes):
  sns.boxplot(data=df, x=col, ax=ax)
plt.show()
```

* viloinplot 으로 연속 자료형 기술 통계 시각화하기

```python
fig, axes = plt.subplots(1, len(nums), figsize=(10,4))
axes = axes.flatten()
for col, ax in zip(nums, axes):
  sns.violinplot(data=df, x=col, ax=ax)
plt.show()
```

## 속성간 관계 파악
---

### **| 상관분석**

* `corr`를 이용해 상관계수 구하기

```python
df.corr()
```

* `style.background_gradient` 이용해 상관계수 시각화 하기

```python
df.corr().style.background_gradient(cmap='coolwarm')
```

* `heatmap` 이용해 상관계수 시각화 하기

```python
sns.heatmap(data=df.corr(), annot=True, cmap='coolwarm', vmax=1, vmin=-1)
plt.show()
```

* 상관관계가 높은 속성 정리하기




```python
# 성인남성은 많이 죽었다
# 혼자 탄 경우 성인 남성일 확률이 높다
```

### **| 속성간 관계 시각화**

* `pairplot` 이용해 시각화 하기 

```python
# pairplot 은 시간이 오래 걸리기 때문에 일부 샘플을 추출해 그리는게 좋습니다.
sns.pairplot(df.sample(100))
plt.show()
```

* `displot` 이용해 연속 자료형 별 예측값의 분포 `kde` 선으로 시각화 하기 

```python
for col in nums:
  sns.displot(data=df, x=col, kind='kde', hue=y)
plt.show()
```

* 위에서 그린 플롯에 `col` 옵션으로 `sex` 추가하기

```python
for col in nums:
  sns.displot(data=df, x=col, kind='kde', hue=y, col='sex')
plt.show()
```

* 위에서 그린 플롯에 `hue` 옵션으로 `class` 추가하기

```python
for col in nums:
  sns.displot(data=df, x=col, kind='kde', hue='class', col='sex')
plt.show()
```

* `age`과 `fare`의 산점도를 그리고 `hue`옵션에 예측값 추가하기

```python
sns.scatterplot(data=df, x='age', y='fare', hue=y)
plt.show()
```

* `lmplot`을 이용해 위의 산점도를 다시 그리고 `col` 옵션에 `class` 추가하기

```python
sns.lmplot(data=df, x='age', y='fare', hue=y, col='class')
plt.show()
```

* `histplot`으로 범주형 속성들과 예측값 히트맵 그리기

```python
fig, axes = plt.subplots(2, 6, figsize=(24,15))
axes=axes.flatten()
for col, ax in zip(noms, axes):
  sns.histplot(data=df, x=col, y=y, ax=ax)
plt.show()
```
