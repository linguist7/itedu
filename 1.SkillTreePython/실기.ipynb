## 문제3. 점찍기 (10)
---

아래와 같은 출력을 하는 코드를 for 문을 이용해서 작성하세요
```
 *  **********
 **  *********
 ***  ********
 ****  *******
 *****  ******
 ******  *****
 *******  ****
 ********  ***
 *********  **
 **********  *
```

````python
for i in range(0, 11):
  print('*'*i, end="")
  for j in range(6-i):
    print(' ', end="")
  print('*'*((2*j)+1))
````

##**문제5** 클래스 생성  (30)
---

- `seq` 클래스 만들기
- 총 6개의 메서드 구현, 한개당 5점

```
>> a = [1,2,3,4,5]
>> a = seq(a)
>> print(a) # 1번 -> 출력 합니다.
seq([1,2,3,4,5])
>> a.size() # 2번 -> 길이를 반환 합니다.
5
>> a.add(6) # 3번 -> 숫자를 뒤에 추가합니다.
[1,2,3,4,5,6]
>> a.pop(7) # 4번 -> 숫자를 앞에 추가합니다.
[7,1,2,3,4,5]
>> a.double() #5번 -> 모든 값을 제곱합니다.
[1,3,9,16,25]
>> a.tostring() # 6번 -> 모든 값을 문자열에 담아 반환합니다.
"1391625"
```

````python
class seq():
  def __init__(self, num):
      self.num = [num]
  def intro(self):
    print(f'{self.num}')
  def add(self, other):
    return self.num+other.num

a = seq(1)
a.add()
````
