```python
from tensorflow.keras import applications
vgg = applications.VGG16(weights=None, include_top=False, input_shape=(224,224,3))
```

```python
from tensorflow.keras import layers, models, utils
# 전이학습 : transfer learning
def create_transfer(base_model):
  x = layers.Input(shape=(224,224,3))
  y = base_model(x) # y: (4,4,512)
  y = layers.Flatten()(y)
  y = layers.Dense(256, activation='relu')(y)
  y = layers.Dense(5, activation='softmax')(y)
  return models.Model(x, y)
```

```python
model = create_transfer(vgg)
model.summary()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

aug_gen = ImageDataGenerator(rescale=1.0/255.0,
                             rotation_range=40, # 회전
                             width_shift_range=0.2,  # 수평으로 평행이동
                             height_shift_range=0.2,  # 수직으로 평행이동 0-1사이 비율
                             shear_range=0.2, #전단변환
                             zoom_range=0.2, # 랜덤하게 확대할범위
                             horizontal_flip=True, # 수평으로 뒤집기
                             fill_mode='nearest',
                             brightness_range = [0.7,1.3], #밝기?
                             
                             )

gen = ImageDataGenerator(rescale=1.0/255.0)

train_gen = gen.flow_from_directory('/content/new_ds/train',
                                        target_size=(224,224), #H,W만 써준다
                                        batch_size=20,
                                        class_mode='categorical' 
                                        )

valid_gen = gen.flow_from_directory('/content/new_ds/valid',
                                        target_size=(224,224), #H,W만 써준다
                                        batch_size=20,
                                        class_mode='categorical' 
                                        )
```

```python
from tensorflow.keras import losses
```

```python
model.compile(loss = losses.CategoricalCrossentropy(2) ,
              optimizer = 'adam',
              metrics=['acc']
              )

model.fit(train_gen, steps_per_epoch=4, epochs=30, validation_data = valid_gen, validation_steps=2)
```
