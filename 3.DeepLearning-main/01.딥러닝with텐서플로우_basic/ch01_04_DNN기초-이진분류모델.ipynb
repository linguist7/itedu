### | 데이터 준비

#### **데이터 로드**
* `sonar.csv`

````python
data_path = '/content/SkillTreePython-DeepLearning/dataset/sonar.csv'
df = pd.read_csv(data_path, header=None)
df.head()
````

#### **데이터 전처리**

````python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split

def path2data_sonar(path, seed=1):
  # 데이터 적절히 불러오기
  df = pd.read_csv(data_path,
                   header=None, #첫번째 행이 데이터(즉 컬럼이 없음)
                   ) 

 # x-y 분할
  x = df.values[:, :-1] # 모든행(샘플), 0-59까지 열(속성)
  y = df.values[:, -1] # 모든행(샘플), 60열(속성)

 # 정규화- 스탠다드(x의 모든속성, y는 하지 않음)
  scaler = StandardScaler()
  x = scaler.fit_transform(x)
  
 # 정규화- 라벨인코더 (y 라벨링, x는 하지 않음)
  labeling = LabelEncoder()
  y= labeling.fit_transform(y)

 # train-test 데이터 분할
  x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=seed,
                                                      stratify=y) # train-test 분할시 클래스 비율 비슷하게 유지
  return x_train, x_test, y_train, y_test

data_path = '/content/SkillTreePython-DeepLearning/dataset/sonar.csv'
x_train, x_test, y_train, y_test = path2data_sonar(data_path)
print( x_train.shape, x_test.shape, y_train.shape, y_test.shape)
print(x_train[1], y_train[1])

````

* 예측값 분포 확인

````python
plt.subplot(121);sns.countplot(x=y_train);plt.title('trainset')
plt.subplot(122);sns.countplot(x=y_test);plt.title('testset')
````

#### **모델 생성**

````python
from tensorflow.keras import models, layers, utils

def build_model():
  x = layers.Input(shape=(60,)) # 데이터가 주어지면 인풋의 shape는 고정
  z = layers.Dense(30, activation='relu')(x)
  y = layers.Dense(1, activation='sigmoid')(z) # 데이터가 주어지면 아웃풋의 shape는 고정(분류문제는 마지막 아웃풋 shape가 1, 마지막 활성화함수는 시그모이드 함수)
  model = models.Model(x, y, name='sonar_classfier')
  return model
````

#### **모델 시각화**

* `model.summary()`
* `utils.plot_model()`

````python
model = build_model()
model.summary()
utils.plot_model(model, show_shapes=True)
````

### | 학습

#### **모델 컴파일**

```
model.compile(loss, optimizer, metrics) 
```
* `loss` : 'binary_crossentropy'
* `optimizer` : 'adam'
* `metrics`: 'accuracy'


````python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
````

#### **학습**

```
model.fit(x_train, y_train, epochs, batch_size)
```
* `x_train` : 넘파이 혹은 텐서 형식의 인풋 데이터
* `y_train` : 넘파이 혹은 텐서 형식의 아웃풋 데이터
* `epochs` : 학습 횟수
* `batch_size` : 배치 사이즈 (업데이트 한번에 사용될 샘플 개수)

````python
model.fit(x_train, y_train, epochs=100, batch_size=16)
````

### | 검증 및 예측


#### **검증**

```
model.evaluate(x_test, y_test)
```
* `x_test` : 넘파이 혹은 텐서 형식의 인풋 테스트 데이터
* `y_test` : 넘파이 혹은 텐서 형식의 아웃풋 테스트 데이터

````python
model.evaluate(x_test, y_test)
````

#### **예측**

```
model.predict(x_test)
```
* `x_test` : 넘파이 혹은 텐서 형식의 인풋 테스트 데이터
* `y_test` : 넘파이 혹은 텐서 형식의 아웃풋 테스트 데이터

````python
y_pred = model.predict(x_test).flatten()
print(y_pred[0], y_test[0])
y_pred.shape, y_test.shape
````

* 예측 시각화

````python
from sklearn.metrics import confusion_matrix, confusion_matrix, ConfusionMatrixDisplay

# 함수 인풋 배열은 정수가 되야 된다.
cm = confusion_matrix(y_test, np.round(y_pred)) #반올림 (0.5 기준으로)
ConfusionMatrixDisplay(cm).plot(cmap='Blues')
````

````python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras import models, layers, utils

def path2data_sonar(path, seed=1):
  # 데이터 적절히 불러오기
  df = pd.read_csv(data_path,
                   header=None, #첫번째 행이 데이터(즉 컬럼이 없음)
                   ) 

 # x-y 분할
  x = df.values[:, :-1] # 모든행(샘플), 0-59까지 열(속성)
  y = df.values[:, -1] # 모든행(샘플), 60열(속성)

 # 정규화- 스탠다드(x의 모든속성, y는 하지 않음)
  scaler = StandardScaler()
  x = scaler.fit_transform(x)
  
 # 정규화- 라벨인코더 (y 라벨링, x는 하지 않음)
  labeling = LabelEncoder()
  y= labeling.fit_transform(y)

 # train-test 데이터 분할
  x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=seed,
                                                      stratify=y) # train-test 분할시 클래스 비율 비슷하게 유지
  return x_train, x_test, y_train, y_test



def build_model():
  x = layers.Input(shape=(60,)) # 데이터가 주어지면 인풋의 shape는 고정
  z = layers.Dense(30, activation='relu')(x)
  z = layers.Dense(70, activation='relu')(z)
  z = layers.Dense(15, activation='relu')(z)
  y = layers.Dense(1, activation='sigmoid')(z) # 데이터가 주어지면 아웃풋의 shape는 고정(분류문제는 마지막 아웃풋 shape가 1, 마지막 활성화함수는 시그모이드 함수)
  model = models.Model(x, y, name='sonar_classfier')
  return model


data_path = '/content/SkillTreePython-DeepLearning/dataset/sonar.csv'
x_train, x_test, y_train, y_test = path2data_sonar(data_path)
print( x_train.shape, x_test.shape, y_train.shape, y_test.shape)
print(x_train[1], y_train[1])

model = build_model()
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
  
model.fit(x_train, y_train, epochs=100, batch_size=16)
````

````python
y_pred = model.predict(x_test).flatten()

from sklearn.metrics import confusion_matrix, confusion_matrix, ConfusionMatrixDisplay

# 함수 인풋 배열은 정수가 되야 된다.
cm = confusion_matrix(y_test, np.round(y_pred)) #반올림 (0.5 기준으로)
ConfusionMatrixDisplay(cm).plot(cmap='Blues')
````

````python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras import models, layers, utils

def path2data_pima(path, seed=1):
  df = pd.read_csv(data_path, header=None, names = ["pregnant", "plasma", "pressure", "thickness", "insulin", "BMI", "pedigree", "age", "class"])
  x = df.values[:, :-1]
  y = df.values[:, -1] 
  scaler = StandardScaler()
  x = scaler.fit_transform(x)
  x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=seed, stratify=y) 
  return x_train, x_test, y_train, y_test

data_path = '/content/SkillTreePython-DeepLearning/dataset/pima-indians-diabetes.csv'
x_train, x_test, y_train, y_test = path2data_pima(data_path)
print( x_train.shape, x_test.shape, y_train.shape, y_test.shape)
print(x_train[1], y_train[1])
plt.subplot(121);sns.countplot(x=y_train);plt.title('trainset')
plt.subplot(122);sns.countplot(x=y_test);plt.title('testset')
````
