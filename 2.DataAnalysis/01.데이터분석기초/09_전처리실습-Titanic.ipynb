**10. 탑승 위치('Embarked') 별 탑승객 분표 확인하기**

**11. 'Embarked' 필드의 결측치는 값이 가장 많은 'S'로 할당하기**

**※**fillna 함수는 데이터의 비어 있는 값에 인수값을 채울 수 있다.

```python
train['Embarked'] = train['Embarked'].fillna('S') #fillna 함수를 통해 Embarked열의 비어있는 값을 'S'값으로 채움
```

**12. 'Age' 필드의 결측치는 값이 중간값으로 할당하기**

```python
train['Age'] = train['Age'].fillna(train['Age'].median()) #fillna 함수를 통해 Age열의 비어있는 값을 median 함수를 활용해 Age열의 중앙값으로 채움
```

**13. 'Name'필드에서 신분을 나타내는 단어를 뽑아서 'Title' 필드에 할당하기**

**※**extract함수는 pandas에서 정규식표현을 사용할 수 있게 해준다.

```python
train['Title'] = train['Name'].str.extract('([A-Za-z]+)\.', expand=True) #extract함수를 통해 name열의 데이터에서 영문에 '.'으로 되어 있는 부분만 추출해서 'train['Title']' 열에 추가
train.head(10)
```

**14. 'Title' 필드의 카테고리를 6개의 항목으로 변경하기**

**※**replace함수는 데이터 값의 문자열을 교체할 수 있다. 첫번째 인수값은 기존의 문자열이고 두번째 인수값이 변경할 문자열이다.

```python
train['Title'] = train['Title'].replace(['Capt', 'Col', 'Major', 'Dr', 'Rev'], 'Officer') #replace 함수를 통해 Title열의 'Capt', 'Col', 'Major', 'Dr', 'Rev' 값을 'Officer'로 문자열 교체
train['Title'] = train['Title'].replace(['Jonkheer', 'Master'], 'Master') #replace 함수를 통해 Title열의 'Jonkheer', 'Master' 값을 'Master'로 문자열 교체
train['Title'] = train['Title'].replace(['Don', 'Sir', 'the Countess', 'Lady', 'Dona'], 'Royalty')  #replace 함수를 통해 Title열의 'Don', 'Sir', 'the Countess', 'Lady', 'Dona' 값을 'Royalty'로 문자열 교체
train['Title'] = train['Title'].replace(['Mme', 'Ms', 'Mrs'], 'Mrs') #replace 함수를 통해 Title열의 'Mme', 'Ms', 'Mrs' 값을 'Mrs'로 문자열 교체
train['Title'] = train['Title'].replace(['Mlle', 'Miss'], 'Miss') #replace 함수를 통해 Title열의 'Mlle', 'Miss' 값을  'Miss'로 문자열 교체
train['Title'] = train['Title'].replace(['Mr'], 'Mr') #replace 함수를 통해 Title열의 'Mr' 값을 'Mr'로 문자열 교체
```

**16. 변수 y를 선언해서 학습할 목표변수(=종속변수)인 'Survived'필드값을 담기**

```python
y = train.Survived #Survived 열을 y로 바인딩
y
```

**17. 나이('Age') 필드를 그룹핑하여 'AgeGroup'필드 생성하여 할당하기**

**※**pd.cut 함수는 데이터의 범주를 설정합니다.

**※**pd.cut의 첫번째 인수값에 범주를 설정할 데이터 열을 입력하고 두번째 인수값에 범주범위를 입력하고 세번째 인수값에 범주명을 입력한다.

```python
bin = [0, 18, 25, 35, 60, 100]
group_names = ['Baby', 'Youth', 'YoungAdult', 'MiddleAged', 'Senior']
train['AgeGroup'] = pd.cut(train['Age'], bins=bin, labels=group_names) #pd.cut함수를 통해 'Age'열의 bin 범위의 숫자를 grop_names로 범주 설정하여 AgeGroup열에 추가
train['AgeGroup'].value_counts() #AgeGroup열의 데이터의 값을 카운트
```
