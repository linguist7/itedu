### **(3) 마스킹**
---

마스킹이란 넘파이의 불리언 인덱싱과 같은 개념으로 보시면 됩니다. 원하는 조건에 맞는 데이터만 출력할 수 있습니다.

#### **| 인덱싱을 이용한 마스킹**



열 `safety`에서 값이 `A-b`인 마스크를 바인딩 합니다.

```

```



````python
mask = df['safety'] == 'A-b'
mask
````

마스크를 이용해 해당하는 데이터만 출력합니다.
```

df_masked.head()
```

````python
df_masked = df[mask]
df_masked.head()
````

두가지 조건을 다음과 같이 줄 수 있습니다.

```

```

````python
mask = (df['safety']=='A-b') | (df['safety']=='A-a')
df_masked = df[mask]
df_masked.head()
````

#### **| `isin`을 이용한 마스킹**

````python
mask = df['safety'].isin(['A-b'])
mask
````

````python
df_masked = df[mask]
df_masked.head()
````

**예제 01**

다음과 같이 연도, 월, 날짜가 기입된 데이터프레임을 작성하고 `df01` 로 바인딩하세요.

2001년 1-1일 부터 2001년 2-28일까지 기입되어 있습니다. (31+28= 총 59줄)


| index | 연도 | 날짜 | 
|--|--|--|
|0|2001|1-1|
|1|2001|1-2|
|2|2001|1-3|
||...|

````python
a = np.arange(59)
a1 = a.reshape((59,1))
df = pd.DataFrame(a1)

df['연도'] = 2001
month_1 = [f'1-{i}' for i in range(1, 32)]
month_2 = [f'2-{i}' for i in range(1, 29)]
df['날짜'] = month_1 + month_2
df01 = df.drop(0, axis=1)

df01
````

**예제 02**

`df01`의 연도를 모두 2022로 바꾸세요.

````python
df01['연도'] = 2002
df01
````

**예제 03**

`df01` 에 1월, 2월에 이어서 3월을 추가하세요.

````python
b = np.arange(30)
b1 = b.reshape((30,1))
df1 = pd.DataFrame(b1)

df1['연도'] = 2002
month_3 = [f'3-{i}' for i in range(1, 31)]
df1['날짜'] = month_3
df02 = df1.drop(0, axis=1)

df01 = pd.concat([df01,df02], axis=0, ignore_index=True)
df01
````

**예제 06**

`df01`에 `요일` 열을 추가하세요. 첫번째 행의 시작은 `토요일` 입니다.

**예제 09**
