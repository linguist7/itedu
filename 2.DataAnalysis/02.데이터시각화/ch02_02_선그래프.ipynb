### **01. 플롯 분할**
---

````python
x = np.linspace(-5, 5, 1000) #-5부터 5까지 1000개의 데이터
def func2(a):
  y= a*(x**2)
  return y
#왼쪽
plt.subplot(1, 2, 1) #행의 개수, 열의 개수, 위치
plt.plot(x, func2(1)) # y=x^2
plt.plot(x, func2(2)) # y=2x^2


#오른쪽
plt.subplot(1, 2, 2)
plt.plot(x, func2(-1)) #y=-x^2
plt.plot(x, func2(-2)) # y=-2x^2


plt.show()
````

### **02. 눈금 공유**
---



````python
t = np.arange(0.01, 5, 0.01)
s1 = np.sin(t*np.pi)
s2 = np.sin(t**np.pi)
s3 = np.exp(-t)

ax1 = plt.subplot(3, 1, 1)
plt.plot(t, s1)
#눈금 지우기
plt.tick_params('x', labelbottom=False)

plt.subplot(3, 1, 2, sharex=ax1)
plt.plot(t, s2)
plt.tick_params('x', labelbottom=False)

plt.subplot(3, 1, 3, sharex=ax1) #축공유
plt.plot(t, s3)



plt.show()
````

### **01. 로그함수와 지수함수 그리기**
---

````python
x= np.linspace(0.001, 5, 1000)
y1= np.exp(x)
y2= np.exp2(x)
y3= np.log(x)
y4= np.log2(x)
y5= np.log10(x)

#plt.subplot(5, 1, 1) #행의 개수, 열의 개수, 위치
plt.plot(x, y1, label='exp')
plt.plot(x, y2, label='exp2')
plt.legend()
plt.show()
plt.plot(x, y3, label='log')
plt.plot(x, y4, label='log2')
plt.plot(x, y5, label='log10')
plt.legend(loc='best', ncol=1, frameon=True, shadow=True)

plt.show()
````

### **03. flight 데이터**

````python
df = sns.load_dataset('flights')
df.head()
````

````python
plt.plot(df['passengers'])
plt.show()
````

````python
a = df['passengers'].groupby(df['year']).sum()
b = df['passengers'].groupby(df['month']).sum()
print(a, b)
````

````python
#연도별 승객수
plt.plot(a, marker='o', markersize=10, label='Passengers by year', color='C1')
plt.legend(loc='best', fontsize=10, frameon=True, shadow=True)

plt.xlabel('Year',  fontdict={'family': 'serif', 'weight': 'bold'})
plt.ylabel('Passengers',  fontdict={'family': 'serif', 'weight': 'bold'})

plt.tick_params(axis='x', direction='in', length=10, pad=10, labelsize=10, width=5)   
plt.tick_params(axis='y', direction='in', length=10, pad=10, labelsize=10, width=5) 

plt.grid(True)

plt.show()

#월별 승객수
plt.plot(b, marker='^', markersize=10, label='Passengers by month', color='C3')
plt.legend(loc='best', fontsize=10, frameon=True, shadow=True)

plt.xlabel('Month',  fontdict={'family': 'serif', 'weight': 'bold'})
plt.ylabel('Passengers',  fontdict={'family': 'serif', 'weight': 'bold'})

plt.tick_params(axis='x', direction='inout', length=10, pad=10, labelsize=10, width=5) 
plt.tick_params(axis='y', direction='inout', length=10, pad=10, labelsize=10, width=5) 

plt.grid(True, axis='y')

plt.show()
````
