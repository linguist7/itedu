### (1) `plt.hist`의 `histtypes` 옵션비교

```
x1 = np.random.normal(0, 1, 5000)
x2 = np.random.normal(0, 3, 5000)
x3 = np.random.normal(10, 3, 5000)

histtypes=['bar', 'barstacked', 'step', 'stepfilled']

```

````python
x1 = np.random.normal(0, 1, 5000)   # 평균 0 , 편차1
x2 = np.random.normal(0, 3, 5000)   # 평균 0 , 편차3
x3 = np.random.normal(10, 3, 5000)  # 평균 10 , 편차3

histtypes=['bar', 'barstacked', 'step', 'stepfilled']
````

````python
#2x2 플롯
for i, histtype in enumerate(histtypes):
  plt.subplot(2,2,i+1)
  plt.hist([x1,x2,x3], histtype=histtype, label=['x1', 'x2', 'x3'], alpha=0.5, bins=50)
  plt.legend()
  plt.title(histtype)


plt.show()

````

### (2) 객체지향 사용하기

```
fig, ax = plt.subplots(2,2, figsize=size)
```

````python
fig, ax = plt.subplots(2,2, figsize=(8,6))
ax = ax.flatten()#2차원을 1차원배열로 바꿈 2x 2이상만

for i, histtype in enumerate(histtypes):
  ax[i].hist([x1,x2,x3], histtype=histtype, 
             label=['x1', 'x2', 'x3'], 
             alpha=0.5, 
             bins=50)
  ax[i].legend()
  ax[i].set_title(histtype)


plt.show()
````

### (3) Iris 데이터 변수별 분포 그리기

```
           꽃받침 길이   꽃받침 너비   꽃의 길이  꽃의 너비

```


```
from sklearn.datasets import load_iris

```


* `.feature_names` : 변수 이름
* `.target_names` : 타겟 이름
* `.data` : 데이터 
* `.target` : 타겟


````python
from sklearn.datasets import load_iris
iris = load_iris()
````

````python
 # 변수이름/ 라벨이름
x_names = iris.feature_names
y_names = iris.target_names
print('x_name', x_names)
print('y_name', y_names)
````

````python
# 데이터셋
x_data = iris.data
y_data = iris.target
print('x', x_data.shape)
print('y', y_data.shape)
````

````python
plt.hist(y_data)
plt.xticks([0,1,2], y_names)
plt.title('Numberof each target')
plt.show()
````

````python
#x_data.shape
plt.figure(figsize=(6,3))
plt.hist(x_data, histtype='stepfilled', bins=30, alpha=.5, label=x_names) #2차원배열
plt.legend()
plt.show()
````

### (1)  지수분포(exponential distribution) 그리기




````python
a = np.random.exponential(scale=0.5, size=10000)
b = np.random.exponential(scale=1, size=10000)
c = np.random.exponential(scale=2, size=10000)
````

* 지수분포의 확률밀도 함수는 다음과 같습니다.


![](https://wikimedia.org/api/rest_v1/media/math/render/svg/7ae26c8e2d08abfda4102c31e1f380e7fa9068b7)

