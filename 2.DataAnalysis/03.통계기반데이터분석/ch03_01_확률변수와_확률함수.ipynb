#### **예제 01. 공정한 주사위**



* 확률변수가 '주사위를 굴려 나온 눈의 값' 일 때, 확률변수 $X$를 정의하세요.

```python
X = [ 1,2,3,4,5,6 ] #벡터 리스트 어레이 아무거나
```

* 주사위의 모든 면이 '공정' 할 때, 확률 질량 함수 $P$를 정의하세요.

```python
def P(x):
  if 1<= x <=6: return 1/6
  else: 0
  
P_X = list(map(P, X))
print(P_X)
```

* 확률 분포를 그리세요.

```python
plt.title('pmf')
_=plt.bar(X, P_X)
```

* 누적 확률 분포를 그리세요.

```python
F_X = np.cumsum(P_X)
print(F_X)
plt.title('cdf') #누적 밀도함수 -> 연속형 분포 
_=plt.plot(F_X, marker='o')
```

```python
_=plt.plot(np.cumsum(P_x), marker='o')
```

#### **예제 03. 화재 발생**

#### **예제 04. 공정하지 않은 주사위**

* 주사위의 짝수 면이 나올 확률이 홀수 면이 나올 확률의 2배일 때, 확률 질량 함수 $P$를 정의하세요.

```python
_=plt.plot(np.cumsum(P_x), marker='o')
```

#### **예제 05. 주사위 두번 굴리기**


* 확률변수가 '주사위를 두번 굴려 나온 눈의 합' 일 때, 확률변수 $X$를 정의하세요.

```python
S = []
for i in range(1,7):
  for j in range(1,7):
    S.append([i,j])
print(S)

X=set()
for s in S:
  X.add(sum(s))
X = list(X)
print(X)
```

*  확률 질량 함수 $P$ 를 정의하세요.

```python
P=dict()
for s in S:
  if sum(s) in P: P[sum(s)]+=1/36
  else: P[sum(s)]=1/36
print(P)
```

* 확률 분포를 그리세요.

```python
P_x = list(map(lambda x: P[x], X))
print(P_x)
_=plt.bar(X, P_x)
```

* 누적 확률 분포를 그리세요.

```python
_=plt.plot(np.cumsum(P_x), marker='o')
```

#### **예제 06. 손의 길이**


* 확률변수가 '성인의 손 길이(cm)' 입니다. $X$를 10부터 20까지 1000개의 값으로 정의하세요.

```python
X = np.linspace(10, 20, 1000)
X
```

* 손 길이에 대한 확률이 다음과 같이 주어졌을 때 함수 $f$ 를 정의하세요. ($\sigma$ = 1)



\begin{align}
 f(x)=\frac{1}{\sqrt{2\pi}\sigma}\exp^{-\frac{(x-15.2)^2}{2*\sigma^2}}
\end{align}

```python
def f(x, sigma=1):
  return 1/np.sqrt(2*np.pi*sigma)*np.exp(-1*((x-15.2)**2)/2*sigma**2)
```

* 확률 분포를 그리세요.

```python
f_X = f(X)
_=plt.plot(X, f_X)
```

* 누적 확률 분포를 그리세요.

```python
_=plt.plot(X, np.cumsum(f_X))
```

#### **예제 07. 파라미터 변경**

예제 06 에서 정의한 함수 $f$의 파라미터 $\sigma$를 각각 `0.5, 1, 2, 4`로 정의한 후 확률분포와 누적확률 분포를 그리세요.

\begin{align}
 f(x)=\frac{1}{\sqrt{2\pi}\sigma}\exp^{-\frac{(x-15.2)^2}{2*\sigma^2}}
\end{align}

```python
fig, axes = plt.subplots(2,4, figsize=(15,6))
axes = axes.flatten()
for i, sigma in enumerate([0.5,1,2,4]):
  axes[i].plot(X, f(X,sigma=sigma))
  axes[i].set_title(f'sigma : {sigma}')  
  axes[i+4].plot(np.cumsum(f(X,sigma=sigma)))
```
