### 일변수 스칼라 함수
---

하나의 인풋(=스칼라, 일변수), 하나의 아웃풋(=스칼라, 스칼라함수)

\begin{align}
y = f(x)
\end{align}

#### **예제 01**

\begin{align}
f(x) = x^2+2x+1
\end{align}

* 함수 생성하기

````python
def f1(x): #x=스칼라
  return (x**2)  +  (2*x)  +  1

f1(-1)
````

* 인풋 정의 (-10~10, 100개)

````python
# 100개의 벡터
xv = np.linspace(-10,10,100)
# 100개 스칼라 값에 대한 100개 스칼라 아웃풋
fv = f1(xv)
xv.shape, fv.shape
````

* 2d플롯으로 시각화

````python
fig, ax = plt.subplots(1)
ax.plot(xv,fv)
ax.set_title('f(x) according to x')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

x = -1
ax.plot(x, f1(x), marker='o', c='r')
ax.text(x, f1(x), f'({x},{np.round(f1(x),2)})')   #np.round: 소수점 아래 2번째까지
plt.show()
````

* 그래프 설명
  * x가 가질수 있는 값 : 모든 실수
  * f(x)가 가질 수 있는 값: 0이상 모든 실수
  * 그래프의 특징 : -1일때 아웃풋이 가장 작고, 멀어질때 아웃풋이 커진다.

#### **예제 02**

\begin{align}
f(x) = cosin(x)
\end{align}

* 함수 생성하기

````python
def f2(x):
  return np.cos(x)
````

* 인풋 정의 (-10~10, 100개)

````python
xv = np.linspace(-10,10,100)
fv = f2(xv)
fv.shape
````

* 2d플롯으로 시각화

````python
fig, ax = plt.subplots(1)
ax.plot(xv,fv)
ax.set_title('f(x) according to x')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

x = -6.3
ax.plot(x, f2(x), marker='o', c='r')
ax.text(x, f2(x), f'({x},{np.round(f2(x),2)})')   #np.round: 소수점 아래 2번째까지
plt.show()
````

* 그래프 설명
  * x가 가질수 있는 값 (인풋) : 모든 실수
  * f(x)가 가질 수 있는 값 (아웃풋) : -1 에서 1 사이의 모든 실수
  * 그래프의 특징 : -3.1일때 아웃풋이 가장 작고, 0.1일때 가장크다. 3.2주기별로 같아진다.

### 다변수 스칼라 함수
---

여러개의 인풋(=다변수), 하나의 아웃풋(=스칼라, 스칼라함수)


\begin{align}
y = f(𝐱)\\
\end{align}



#### **예제 03**

\begin{align}
f(x, y) = x+y
\end{align}

* 함수 생성하기

````python
def f3(x,y):
  return x+y

f3(2,2)
````

* 인풋 정의 (x, y, 모두 -15~20 값)

````python
xv=np.linspace(-15,20,36)
yv=np.linspace(-15,20,36)
xv.shape, yv.shape
print(xv.shape)
print(yv.shape)
X, Y = np.meshgrid(xv,yv) # xv, yv의 순서쌍 모두 반환
print(X.shape,Y.shape)
F = f3(X,Y) # X,Y 모든 쌍에 대한 함수 결과 반환
print(F.shape)
````

* 3d플롯으로 시각화

````python
# 3차원 플롯

fig = plt.figure(figsize=(9,7))
ax = plt.axes(projection='3d') #3차원플롯
ax.scatter3D(X, Y, f3(X, Y), marker='.', color='k')
surf = ax.plot_surface(X, Y, f3(X, Y), cmap='rainbow', alpha=.7) # 점을 이은 면 surface 그리기
#설명
ax.set_title('f(x,y) according to x,y')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')

#컬러바
fig.colorbar(surf, label='f(x,y)', shrink=.6)
plt.show()
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : 모든 실수
  * f(x)가 가질 수 있는 값 (아웃풋) : 모든 실수
  * 그래프의 특징 :함께 증가 하면 아웃풋이 함께 커진다. 최소값이 없음 무한

#### **예제 04**

\begin{align}
f(x, y) = (x-5)^2+(y-5)^2 
\end{align}

* 함수 생성하기

````python
def f4(x, y):
  return ((x-5)**2)    +   ((y-5)**2)
````

* 인풋 정의(x, y 모두 -15~20 사이 값)

````python
xv = np.linspace(-15, 20, 36)
yv = np.linspace(-15, 20, 36)

X, Y = np.meshgrid(xv,yv)
f4(xv, yv).shape
````

* 3d플롯으로 시각화

````python
# 3차원 플롯
fig = plt.figure(figsize=(9,7))
ax = plt.axes(projection='3d') #3차원플롯
ax.scatter3D(X, Y, f4(X, Y), marker='.', color='k')
surf = ax.plot_surface(X, Y, f4(X, Y), cmap='rainbow', alpha=.7) # 점을 이은 면 surface 그리기
#설명
ax.set_title('f(x,y) according to x,y')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')

#컬러바
fig.colorbar(surf, label='f(x,y)', shrink=.7)
plt.show()
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : 모든 실수
  * f(x)가 가질 수 있는 값 (아웃풋) : 0보다 큰 모든 실수
  * 그래프의 특징 : 최소값이 0 이다.

#### **예제 05**

\begin{align}
f(x_1, x_2) = 50(x_2-x_1^2)^2+(2-x_1)^2

## 일변수 벡터함수
---

하나의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)




\begin{align}
f(x) = (f_1(x), f_2(x), ..., f_n(x))^T
\end{align}

#### **예제 06**

\begin{align}
f(x) = \begin{pmatrix}
h(x)\\ 
g(x)\\ 
\end{pmatrix}  = \begin{pmatrix}
x^2\\ 
5x\\ 
\end{pmatrix}
\end{align}

* 함수 $f(x)$ 생성하기

````python
def f6(x):
  return np.array([x**2, 5*x])
````

* 인풋 정의 (-10~10 20개)

````python
xv = np.linspace(-10,10,20)
hv, gv = f6(xv)
#인풋   첫번째 벡터요소      : 두번째 벡터요소
xv.shape,     hv.shape,   gv.shape
````

* 2d플롯으로 시각화

````python
#아웃풋(f(x)) 스페이스 그려보기
#x촉을 h(x), y촉을 g(x)

fig, ax = plt.subplots(1, figsize=(6,5))
ax.plot(gv, hv)
#설명
ax.set_title('f(x) space')
ax.set_xlabel('g(x)')
ax.set_ylabel('h(x)')

scat = ax.scatter(gv, hv, c=xv, cmap='rainbow')
fig.colorbar(scat)

plt.show()
````

* 그래프 설명
  * x가 가질수 있는 값 (인풋) : 모든 실수
  * g(x)가 가질 수 있는 값 (첫번째 아웃풋) : 모든 실수
  * h(x)가 가질수 있는 값 (두번째 아웃풋): 0이상 모든 실수

#### **예제 07**

\begin{align}
f(t) = \begin{pmatrix}
x(t)\\ 
y(t)\\ 
z(t)\\ \end{pmatrix} = \begin{pmatrix}
sin(6t)\\
\frac{t}{4}\\
\frac{t^2}{2}\\
\end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f7(t):
  return np.array([np.sin(6*t), t/4, (t**2)/2])
````

* 인풋 정의 (-1~3 200개)

````python
t = np.linspace(-1,3,200)
x, y, z = f7(t)
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize=(9,7))
ax = plt.axes(projection='3d')
scat=ax.scatter3D(x,y,z, c=t,  cmap='rainbow')
ax.set_title('f(t) space')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

#컬러바
fig.colorbar(scat, label='t', shrink=.7)
plt.show()
````

* 그래프 설명
  * t가 가질수 있는 값 (인풋) : 모든 실수
  * x(t)가 가질 수 있는 값 (첫번째 아웃풋) : -1에서 1까지의 모든 실수
  * y(t)가 가질 수 있는 값 (두번째 아웃풋) : 모든실수
  * z(t)가 가질 수 있는 값 (세번째 아웃풋) : 0이상 모든 실수

## 다변수 벡터함수
---
여러개의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)

\begin{align}
f(𝐱) = (f_1(𝐱), f_2(𝐱), ..., f_n(𝐱))^T
\end{align}

#### **예제 09**


\begin{align}
f(x,y) = \begin{pmatrix}
x^2\\ 
y^2\\ 
xy\\ \end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f9(x, y):
  return np.array([x**2, y**2, x*y])

f9(1, 2)
````

* 인풋 정의 (x,y 모두 -0.5 ~ 1 사이 50개)

````python
xv = np.linspace(-0.5,1,50)
yv = np.linspace(-0.5,1,50)


X, Y = np.meshgrid(xv, yv) # xv, yv의 순서쌍 모두 반환
#아웃풋정의
A, B, C = f9(X, Y)
X.shape, Y.shape, A.shape, B.shape, C.shape
````

* 3d플롯으로 시각화

````python
# 3차원 플롯

fig = plt.figure(figsize=(9,7))
ax = plt.axes(projection='3d') #3차원플롯
# 아웃풋 스페이스 (A,B,C)
ax.scatter3D(A, B, C, marker='.', color='k')
ax.plot_surface(A, B, C, alpha=.5) # 점을 이은 면 surface 그리기
#설명
ax.set_title('f(x,y) space')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('xy')
````

#### **예제 10**


\begin{align}
f(u,v) = \begin{pmatrix}
u\\ 
v\\ 
1+u^2+\frac{v}{1+v^2}\\ \end{pmatrix}, \ \ \ \ \ 0 \leq u, v \leq1
\end{align}

* 함수 생성하기

````python
def f10(u,v):
  return np.array([u, v, 1+(u**2)+v/(1+v**2)])
````

* 인풋 정의 (u, v 모두 0 ~ 1 사이 50개)

````python
u = np.linspace(0,1,50)
v = np.linspace(0,1,50)
U, V = np.meshgrid(u,v)
A, B, C = f10(U,V)
````

* 3d플롯으로 시각화

````python
# 3차원 플롯
fig = plt.figure(figsize=(9,7))
ax = plt.axes(projection='3d') #3차원플롯
# 아웃풋 스페이스 (A,B,C)
ax.scatter3D(A, B, C, marker='.', color='k')
ax.plot_surface(A, B, C, alpha=.5) # 점을 이은 면 surface 그리기
#설명
ax.set_title('f(u,v) space')
ax.set_xlabel('u')
ax.set_ylabel('v')
ax.set_zlabel('1+u**2+v/1+v**2')

plt.show()
````
